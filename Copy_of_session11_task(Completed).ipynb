````python
l = [{0:2},2,3,4,'5', {5:10}]
# For calculating sum of above list


s=0

for i in range(len(l)):
    
    try:
        s+=l[i].get(i)
    except AttributeError:
        pass
    except Exception :
        pass
    
    # try:
    #     s+=l[i]
    # except TypeError:
    #     pass
    # except Exception:
    #     pass

    try:
        
        s+=int(l[i])
    except (ValueError,TypeError):
        pass
    except Exception:
        pass

print(s)
````

````python
l = [{0:2}, 2, 3, 4, '5', {5:10}]
s = 0

for i in range(len(l)):
    try:
        # if element is a dict, add its only value
        if isinstance(l[i], dict):
            s += list(l[i].values())[0]
        # if element is a string number
        elif isinstance(l[i], str):
            s += int(l[i])
        # if element is already integer
        elif isinstance(l[i], int):
            s += l[i]
    except (ValueError, TypeError, AttributeError) as e:
        print("Skipped element due to error:", e)

print("Sum of elements:", s)
````

### `Q-3:`: File Handling with Exception handling

Write a program that opens a text file and write data to it as "Hello, Good Morning!!!". Handle exceptions that can be generated during the I/O operations. Do not show the success message on the main exception handling block (write inside the else block).

````python
try:
    f= open('sample.txt','w')
except FileNotFoundError:
    print("File not found")
except Exception as e:
    print("Unexpected error",  e)
else:
    f.write("Hello, Good Morning!!!")
    print('successful!!')
finally:
    f.close()
````

### `Q-4`: Number game program.

Write a number game program. Ask the user to enter a number. If the number is greater than number to be guessed, raise a **ValueTooLarge** exception. If the value is smaller the number to be guessed the, raise a **ValueTooSmall** exception and prompt the user to enter again. Quit the program only when the user enters the correct number. Also raise **GuessError** if user guess a number less than 1.

````python
class ValueTooLarge(Exception):

    def __init__(self):
        print("Value is too large! Please try again!!\n")

class ValueTooSmall(Exception):

    def __init__(self):
        print("Value is too small! Please try again!!\n")
        

class GuessError(Exception):
    def __init__(self):
        print("Value cant be less than 1! Please try again!!\n")
        

import random
n=random.randint(0,100)

def guess_number(num):
    if num<1:
        raise GuessError()
    elif num<n:
        raise ValueTooSmall()
    elif num>n:
        raise ValueTooLarge()
    elif num==n:
        print("Congratulation!! You have sucessfully guessed the number." )
        print("Thanks for playing!!")
        return True


while True:
    try:
        new=int(input("ENter the num between 1 to 100"))
        if guess_number(new):
            break
    except (ValueTooLarge,ValueTooSmall,GuessError) as f:
        pass

    
````

### `Q-5:` Cast vote

Write a program that validate name and age as entered by the user to determine whether the person can cast vote or not. To handle the age, create **InvalidAge** exception and for name, create **InvalidName** exception. The name will be invalid when the string will be empty or name has only one word.

Example 1:

Input:

```bash
Enter the name:               goransh singh
Enter the age: 25
```

Output:

```bash
Goransh Singh  Congratulation !!! You can vote.
```

````python
class InvalidName(Exception):
    def __init__(self):
        print("Invalid Name.Sorry!! You cant vote!! ")

class InvalidAge(Exception):
    def __init__(self):
        print("Invalid Age.Sorry!! You cant vote!! ")        


def vote_validation(name,age):
    if len(name)==0 or len(name)==1:
        raise InvalidName()
    elif age<18:
        raise InvalidAge()
    else:
        print(f"{name} Congrulation !!! You can vote")

try:
    vote_validation('Singh',25)
except (InvalidAge,InvalidName) :
    pass
````

### `Q-6`: Write a python function which infinitely prints natural numbers in a single line. Raise the **StopIteration** exception after displaying first 20 numnbers to exit from the program.

````python
def number():
    i=0
    while True:
        i+=1
        print(i, end=' ')
        
        if i==20:
            raise StopIteration


try:
    number()
except StopIteration:
    pass
````

````python
def number():
    i=1
    while True:
        yield i  #its like return but instead of ending it pauses
        i+=1


gen=number()
try:
    for _ in range(20):
        print(next(gen),end=' ')
    raise StopIteration
except StopIteration:
    print("\n Stopped after 20th iteration")
````
