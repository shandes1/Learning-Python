###`Q-5.`Write a function to change the data type of given a column or a Series. Function takes series and data type as input, returns the converted series.
```
series = pd.Series([1,2,'Python', 2.0, True, 100])
change to float type data
```
Note: Read about pd.to_numeric()

````python
series = pd.Series([1,2,'Python', 2.0, True, 100])
series
def change_datatype(s1,datatype):
    if datatype==float:
        return pd.to_numeric(s1,errors="coerce").astype(float)
    elif datatype==int:
        return pd.to_numeric(s1,errors="coerce").astype(int)
    else:
        raise("cant do it")
    

change_datatype(series,float)
````

````python
new_item=items.str.replace("$","Rs",regex=False)
new_item
````

````python
float_item=new_item.str.replace("Rs"," ",regex=False)
float_item=pd.to_numeric(float_item,errors="coerce")
float_item
````

````python
float_item.fillna(float_item.mean())
````
