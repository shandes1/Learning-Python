### Q9:- Given 2 fractions, find the sum of those 2 fractions.Take the numerator and denominator values of the fractions from the user.

````python
print('Enter your 1st faction')
a=int(input('a= '))
b= int(input('b= '))

print('Enter your 2nd faction')
c= int(input('c= '))
d= int(input('d= '))

denominator=int(b*d)
numerator=int(a*d+c*b)

from math import gcd
common=gcd(numerator,denominator)

do=denominator// common
n=numerator// common

if do==1:
 print(f"The value of {a}/{b} + {c}/{d} = {n/do}")
else:
 print(f"The value of {a}/{b} + {c}/{d} = {n}/{do}")
````

### Q10:- Given the height, width and breadth of a milk tank, you have to find out how many glasses of milk can be obtained? Assume all the inputs are provided by the user.



Input:<br>
Dimensions of the milk tank<br>
H = 20cm, L = 20cm, B = 20cm
<br><br>
Dimensions of the glass<br>
h = 3cm, r = 1cm

````python
import math
print('Enter the height,length and breadth  of milk tank:')
l=float(input('Length= '))
b=float(input('Breadth= '))
h=float(input('Height= '))

print('Enter dimensions of glass: ')
H=float(input('Height= '))
r=float(input('Radius= '))

VOLUME=l*b*h
volume= math.pi * r**2 *H

n=int(VOLUME/volume)
print("No of glass:",n)
````
