### **`Problem-1:`** Write a Python function that takes a list and returns a new list with unique elements of the first list.

**Exercise 1:**

Input:

```bash
[1,2,3,3,3,3,4,5]
```

Output:

```bash
[1, 2, 3, 4, 5]
```

````python
def new_lis(l):
  new_set=set(l)
  new_list=list(new_set)
  return new_list

l=[1,2,3,3,3,3,4,5]
l1=new_lis(l)
print(l1)
````

### **`Problem-2:`** Write a Python function that accepts a hyphen-separated sequence of words as parameter and returns the words in a hyphen-separated sequence after sorting them alphabetically.

**Example 1:**

Input:
```bash
green-red-yellow-black-white
```

Output:
```bash
black-green-red-white-yellow
```

````python
string='green-red-yellow-black-white'
def str_sort(string):
  l=[]
  new=''
  for i in string:
    if i != '-':
      new=new+i
    elif i=='-':
      l.append(new)
      new=''
  l.append(new)
  l=sorted(l)
  for i in l:
    new_string='-'.join(l)
  print(new_string)

str_sort(string)
````

````python
#short version
string='green-red-yellow-black-white'
def str_sort(string):
  l= string.split('-')
  l=sorted(l)
  for i in l:
    new_string='-'.join(l)
  print(new_string)

str_sort(string)
````

### **`Problem 3:`** Write a Python function that accepts a string and calculate the number of upper case letters and lower case letters.

```
Sample String : 'CampusX is an Online Mentorship Program fOr EnginEering studentS.'
Expected Output :
No. of Upper case characters :  9
No. of Lower case Characters :  47
```

````python
def uppercase(string):
  upper=0
  lower=0
  for i in string:
    if i.isupper():
      upper +=1
    elif i.islower():
      lower +=1
  print(f'No of upper case characters : {upper}')
  print(f'No of lower case characters : {lower}')

String = 'CampusX is an Online Mentorship Program fOr EnginEering studentS.'
uppercase(String)
````

### **`Problem 5:`** Write a Python function to check whether a number is perfect or not.

A Perfect number is a number that is half the sum of all of its positive divisors (including itself).

Example :

```
The first perfect number is 6, because 1, 2, and 3 are its proper positive divisors, and 1 + 2 + 3 = 6.
Equivalently, the number 6 is equal to half the sum of all its positive divisors: ( 1 + 2 + 3 + 6 ) / 2 = 6.

The next perfect number is 28 = 1 + 2 + 4 + 7 + 14. This is followed by the perfect numbers 496 and 8128.
```


````python
def perfect(num):
  #show divisors
  l1=[ x for x in range(1,num+1) if num%x==0]
  print(l1)
  #value of sum and half of sum
  sum=0
  half_of_sum=0
  for i in l1:
    if i<num:
      sum+=i
    half_of_sum+=i
  #compare
  if sum==num and half_of_sum/2==num:
    print(f'{num} is a perfect number')
  else:
    print(f'{num} is not a perfect number')

perfect(8128)
````

`Problem-7` Write a python function that accepts a string as input and returns the word with most occurence.

```
Input:
hello how are you i am fine thank you
```

```
Output
you -> 2
```

````python
def most_occurance(string):
  new_dic={}
  words=string.split()
  for i in words:
    if i in new_dic:
      new_dic[i]+=1
    else:
      new_dic[i]=1
  max_word= max(new_dic,key=new_dic.get)
  print(f'{max_word}-> {new_dic[max_word]}')

string='hello how are you i am fine thank you'
most_occurance(string)
````

`Problem-8` Write a python function that receives a list of integers and prints out a histogram of bin size 10

```
Input:
[13,42,15,37,22,39,41,50]
```

```
Output:
{11-20:2,21-30:1,31-40:2,41-50:3}
```

````python
def histogram(data):
    bins = {}

    for num in data:
        bin_start = (num // 10) * 10 + 1  # Bin starts at 1, 11, 21, etc.
        bin_end = bin_start + 9  # Bin ends at 10, 20, 30, etc.
        bin_label = f"{bin_start}-{bin_end}"  # Format as "11-20"

        bins[bin_label] = bins.get(bin_label, 0) + 1  # Count occurrences

    # Sorting bins by numerical start value
    sorted_bins = dict(sorted(bins.items(), key=lambda x: int(x[0].split('-')[0])))

    # Formatting output as required
    output = "{" + ",".join(f"{k}:{v}" for k, v in sorted_bins.items()) + "}"

    print(output)  # Print output in required format

# Example usage
numbers = [13, 42, 15, 37, 22, 39, 41, 50]
histogram(numbers)
````

`Problem-9` Write a python function that accepts a list of 2D co-ordinates and a query point, and then finds the the co-ordinate which is closest in terms of distance from the query point.

```
List of Coordinates
[(1,1),(2,2),(3,3),(4,4)]
Query Point
(0,0)
```

```
Output
Nearest to (0,0) is (1,1)
```

````python
import math
def  distance(l,point):
  dic={}
  for i in l:
    distance=(i[0]-point[0])**2+(i[1]-point[1])**2
    result=math.sqrt(distance)
    dic[i]=result

  sorted_dic = dict(sorted(dic.items(), key=lambda item: item[1]))
  nearest_point = list(sorted_dic.items())[0]
  print(f'Nearest point to {point} is {nearest_point[0]}')

distance([(1,1),(2,2),(3,3),(4,4)],(0,0))
````

`Problem-10`:Write a python program that receives a list of strings and performs bag of word operation on those strings

https://en.wikipedia.org/wiki/Bag-of-words_model

````python
def bag_of_words_fun(texts):

  for text in texts:
    words=text.split()
    bag_of_word={}
    for word in words:
      if word in bag_of_word:
        bag_of_word[word]+=1
      else:
        bag_of_word[word]=1
    print( bag_of_word)


example={
    "I love programming I in Python",
    "Python programming is fun",
    "I enjoy solving problems with Python"
}
bag_of_words_fun(example)
````
